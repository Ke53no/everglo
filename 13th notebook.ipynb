```python
from tkinter import *
from tkinter import ttk
from enum import Enum

class State(Enum):
    calculated = 0
    in_number1 = 1
    post_operator = 2
    in_number2 = 3

class Calculator(Tk):
    def __init__(self, *args, **kwargs):
        super().__init__(*args,**kwargs)
        self.title('Simple Calculator')
        self.geometry('400x600')
        
        self.result_string = StringVar()
        self.result_string.set('0')
        self.number1 ='0'
        self.number2 =None
        self.operator = None
        self.state = State.calculated  
        
        # Output
        output=ttk.Label(self, textvariable=self.result_string, font=('Beleren', 48), anchor = E)
        output.pack(side=TOP, expand=True, fill=BOTH)
                
                #row 1
        row = Frame(self)
        row.pack(side=TOP, expand=True, fill=BOTH)
        self.add_button(row, 'CE', self.on_ce_click)
        self.add_button(row, 'C', self.on_c_click)
        self.add_button(row, 'Back', self.on_back_click)
        self.add_button(row, '/', self.on_division_click)
        
                #row 2
        row = Frame(self)
        row.pack(side=TOP, expand=True, fill=BOTH)
        #self.add_button(row, '7', self.on_7_click)
        self.add_button(row, '7', self.gen_on_number_click('7'))
        #self.add_button(row, '8', self.on_8_click)
        self.add_button(row, '8', self.gen_on_number_click('8'))
        #self.add_button(row, '9', self.on_9_click)
        self.add_button(row, '9', self.gen_on_number_click('9'))
        self.add_button(row, 'x', self.on_multiply_click)
        
                #row 3
        row = Frame(self)
        row.pack(side=TOP, expand=True, fill=BOTH)
        self.add_button(row, '4', self.on_4_click)
        self.add_button(row, '5', self.on_5_click)
        self.add_button(row, '6', self.on_6_click)
        self.add_button(row, '-', self.on_minus_click)
        
                #row 4
        row = Frame(self)
        row.pack(side=TOP, expand=True, fill=BOTH)
        #self.add_button(row, '1', self.on_1_click)
        self.add_button(row, '1', self.gen_on_number_click('1'))
        self.add_button(row, '2', self.on_2_click)
        self.add_button(row, '3', self.on_3_click)
        self.add_button(row, '+', self.on_plus_click)
        
                #row 5
        row = Frame(self)
        row.pack(side=TOP, expand=True, fill=BOTH)
        self.add_button(row, '+/-', self.on_negate_click)
        self.add_button(row, '0', self.on_0_click)
        self.add_button(row, '.', self.on_dot_click)
        self.add_button(row, '=', self.on_equal_click)
            
    def add_button(self, parent, text=None, command=None):
        button = ttk.Button(parent, text = text, command = command)
        button.pack(side=LEFT, expand=True, fill=BOTH)
        
    def on_number(self,ch):
        if self.state==State.calculated:
            self.number1 = ch
            self.state = State.in_number1
            self.result_string.set(self.number1)
        elif self.state == State.in_number1:
            self.number1 = self.number1 + ch
            self.result_string.set(self.number1)
        elif self.state == State.post_operator:
            self.number2 = ch
            self.state == State.in_number2
            self.result_string.set(self.number2)
        elif self.state == State.in_number2:
            self.number2 = self.number2 + ch
            self.result_string.set(self.number2)
    
    def on_operator(self,operator):
        if self.state==State.calculated:
            self.state = State.post_operator
        elif self.state == State.in_number1:
            self.state = State.post_operator
        #elif self.state == State.post_operator:
        elif self.state == State.in_number2:
            self.on_equal_click()
            self.state = State.post_operator
        else: assert(0)
            
        self.operator = operator
    
    #chenge in making buttons, able to replace on_%number%_click funcs
    def gen_on_number_click(self, ch):
        return lambda: self.on_number(ch)
    
    def on_ce_click(self):
        print('ce_clicked')
    def on_c_click(self):
        print('c_clicked')
    def on_back_click(self):
        print('back_clicked')
    def on_division_click(self):
        print('division_clicked')
        self.on_operator('/')
        
    #def on_7_click(self):
    #    print('7_clicked')
    #    self.on_number('7')
    #def on_8_click(self):
    #    print('8_clicked')
    #    self.on_number('8')
    #def on_9_click(self):
    #    print('9_clicked')
    #    self.on_number('9')
    def on_multiply_click(self):
        print('multiply_clicked')
        self.on_operator('*')
        
    def on_4_click(self):
        print('4_clicked')
        self.on_number('4')
    def on_5_click(self):
        print('5_clicked')
        self.on_number('5')
    def on_6_click(self):
        print('6_clicked')
        self.on_number('6')
    def on_minus_click(self):
        print('minus_clicked')
        self.on_operator('-')

    #def on_1_click(self):
    #    print('1_clicked')
    #    self.on_number('1')
    def on_2_click(self):
        print('2_clicked')
        self.on_number('2')
    def on_3_click(self):
        print('3_clicked')
        self.on_number('3')
    def on_plus_click(self):
        print('plus_clicked')
        self.on_operator('+')
        
    def on_negate_click(self):
        print('negate_clicked')
    def on_0_click(self):
        print('0_clicked')
    def on_dot_click(self):
        print('dot_clicked')
        self.on_number('.')
    def on_equal_click(self):
        #if self.state != State.in_number2:
        #    return
        
        print('equal_clicked')
        if self.operator == '+':
            self.number1 = str(float(self.number1)+ float(self.number2))
        elif self.operator == '-':
            self.number1 = str(float(self.number1)- float(self.number2))
        elif self.operator == '*':
            self.number1 = str(float(self.number1)* float(self.number2))
        elif self.operator == '/':
            self.number1 = str(float(self.number1)/ float(self.number2))
            
        self.result_string.set(self.number1)
        self.state = State.calculated
            
            
if __name__ == '__main__':
    # if not imported, then run
    app = Calculator()
    app.mainloop()
```
